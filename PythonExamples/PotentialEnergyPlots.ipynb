# Potential Energy

```python
import numpy as np
from matplotlib import pyplot as plt
```

## Potential Energy of  a Harmonic Oscillator

```python
#defining a parabolic function
def U(x): 
    return (x*x)/2

#setting x
x = np.linspace(-5,5,500)

plt.plot(x,U(x))
plt.title('Potential Energy of a Harmonic Oscillator')
plt.xlabel('Energy')
plt.ylabel('U(x)=1/2 x^2')
plt.grid('True')
plt.savefig('POscillator.jpg')

plt.show()
```

## Molecular Potential Energy
Plotting molecular potential energy using the Leonard-Jones or 12-6 potential

https://en.wikipedia.org/wiki/Lennard-Jones_potential

https://nznano.blogspot.com/2017/11/molecular-dynamics-in-python.html

L-J Potential:
- r is the distance between two interacting particles  
- e is the depth of the potential well
- sigma is the distance at which the partcicle potential energy is 0

Interpretation of the model: Two interating particles repel each other at very close distance, attract each other at moderate distance, and do not interact at infinite distanece
